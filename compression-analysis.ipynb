```python
x = 'Compressive strain %'
y = 'Compressive stress MPa'
xline_intact = r9['Compressive strain %'][350:500]
yline_intact = r9['Compressive stress MPa'][350:500]
xline_washed = d9['Compressive strain %'][200:400] 
yline_washed = d9['Compressive stress MPa'][200:400]

fig, ax = plt.subplots(figsize=(8,6))

ax.hold('on')
# Washed S-S Plot
ax = d9.plot(y=y, x=x, xlim=[0, 40], ylim=[0, 40], label='Washed')
# Washed Young's Modulus
ax.plot(xline_washed, yline_washed, label='Linear + Elastic Region, Washed')
# Intact S-S Plot
ax = r9.plot(secondary_y=True, y=y, x=x, xlim=[0, 40], ylim=[0, 40], label='Intact')
# Intact youngs modulus
ax.plot(xline_intact, yline_intact, label='Linear + Elastic Region, Intact')


lines = ax.left_ax.get_lines() + ax.right_ax.get_lines()
ax.left_ax.set_xlabel('Compressive strain (%)')
ax.left_ax.set_ylabel('Compressive stress (MPa)')


ax.legend(lines, [l.get_label() for l in lines], loc='lower right')
```
